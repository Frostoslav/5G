```python
import matplotlib.pyplot as plt
import math
import numpy as np

m = np.random.uniform(size=(10, 2)) #матрица 10х2 с равномерным распределением

print(m) # вывод матрицы

plt.figure(dpi=100)
plt.plot(m[:,0], m[:,1], 'bo') # отображение точек на графике, первый столбец матрицы - х, второй - у
plt.show
```

```python
<h3> Задание 2 </h3>

Сгенерировали 1000 случайных чисел с нормальным распределением Гаусса
и построили их гистограмму, математическое ожидание, дисперсию.

![распределением Гаусса](https://s0.showslide.ru/s_slide/89587ddbd03594b2c9500c1262c6322d/8f4bbbed-5aaa-438f-a4ae-5c895c4f983b.jpeg)
```

```python
numbers = np.random.normal(0, 1, 1000) # 1000 чисел с нормальным распределение Гаусса
print('mean', np.mean(numbers)) #мат. ожидание
print('var', np.var(numbers)) #дисперсия
plt.hist(numbers, bins=100)
plt.show
```

```python
<h3> Задание 3 </h3>

Сгенерировали случайную точку, равномерно распределенную в квадрате
со стороной а. 
```

```python
a = int(input('Введите сторону а: '))
plt.xlim(0, a - 1) # задали границы квадрата со стороной а
plt.ylim(0, a - 1)
point = np.random.uniform(low= 0, high = (a -1), size=(1, 2)) # матрица 1х2 (координаты точки вписанной в квадрат со стороной а) 
                                                              # с равномерным распределением
plt.plot(point[:,0], point[:,1], 'ro') # отображение точки на графике
plt.figure(dpi=100)
plt.show
```
